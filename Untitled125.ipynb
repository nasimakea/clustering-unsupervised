# Q1. What is hierarchical clustering, and how is it different from other clustering techniques?

# ANS â–¶


- Hierarchical clustering is a clustering technique that organizes data points into a hierarchy of clusters. It differs from other clustering techniques such as K-means in that it creates a tree-like structure of clusters rather than a flat partitioning. To illustrate this with a real dataset and code, we'll use the Iris dataset and Python's scikit-learn library.

```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris=load_iris()
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
iris_df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
iris_scaled=scaler.fit_transform(iris_df)
```

```python
from scipy.cluster.hierarchy import dendrogram , linkage
import matplotlib.pyplot as plt
```

```python
linkage_matrix=linkage(iris_scaled,method='ward')
```

```python
plt.figure(figsize=(10, 6))
dendrogram(linkage_matrix, labels=iris.target_names)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Iris Species')
plt.ylabel('Distance')
plt.show()
```

# Differences from Other Clustering Techniques
- Hierarchy: Hierarchical clustering creates a hierarchy of clusters, whereas methods like K-means produce a flat partitioning of data.

- Number of Clusters: Hierarchical clustering does not require specifying the number of clusters beforehand, unlike K-means which requires the number of clusters as input.

- Cluster Shape: Hierarchical clustering can handle clusters of various shapes and sizes, while K-means tends to create spherical clusters.

- Distance Metric: Hierarchical clustering allows for different distance metrics, such as Euclidean or Manhattan distance, while K-means typically uses Euclidean distance.

- Interpretability: The dendrogram produced by hierarchical clustering provides a visual representation of cluster relationships, aiding in cluster interpretation

# Q2. What are the two main types of hierarchical clustering algorithms? Describe each in brief.

# ANS â–¶

- The two main types of hierarchical clustering algorithms are agglomerative hierarchical clustering and divisive hierarchical clustering.

    - Agglomerative hierarchical clustering starts by considering each data point as a single cluster and then iteratively merges the closest clusters until only one cluster remains. It builds a hierarchy of clusters in a bottom-up fashion.

```python
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
```

```python
x=iris.data
```

```python
agglomerat=AgglomerativeClustering(n_clusters=3)
```

```python
agglomerat.fit(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=agglomerat.labels_,cmap='viridis')

plt.xlabel('sepal length')
plt.ylabel("sepal width")
plt.title('Agglomerative Hierachical clustering')
plt.show()
```

- Divisive Hierarchical Clustering:Divisive hierarchical clustering starts with all data points in a single cluster and then recursively divides clusters until each data point is in its own cluster. It builds a hierarchy of clusters in a top-down fashion.

```python
from sklearn.datasets import load_wine
from sklearn.cluster import AgglomerativeClustering
```

```python
import matplotlib.pyplot as plt
```

```python
wine=load_wine()
x=wine.data
```

```python
divisive=AgglomerativeClustering(n_clusters=1,linkage='complete')
divisive.fit(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=divisive.labels_,cmap='viridis')
plt.xlabel("figure 1")
plt.ylabel("figure 2")
plt.title("Divisive Hierachial clustering")
plt.show()
```

# Q3. How do you determine the distance between two clusters in hierarchical clustering, and what are the common distance metrics used?

# ANS â–¶


- In hierarchical clustering, the distance between two clusters is determined using a distance metric, also known as a linkage criterion. Common distance metrics used in hierarchical clustering include such as Euclidean and Manhattan distance

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram , linkage
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
```

```python
iris_df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
iris_df.head()
```

```python
scaler=StandardScaler()
```

```python
iris_scaled=scaler.fit_transform(iris_df)
```

```python
linkage_method=['ward','single','complete','average']
```

```python
plt.figure(figsize=(15,10))
```

```python
for i, method in enumerate(linkage_method,1):
  plt.subplot(2,2,i)
  linkage_matrix=linkage(iris_scaled,method=method)
  dendrogram(linkage_matrix)
  plt.title(f'Hierarchical Clustering Dendrogram ({method} linkage)')
  plt.ylabel('Distance')
  plt.xlabel('Iris Species')

plt.tight_layout()
plt.show()
```

# ANS â–¶

- Determining the optimal number of clusters in hierarchical clustering involves selecting the point at which the dendrogram best represents the underlying structure of the data. Several methods are commonly used for this purpose, including the elbow method, silhouette score, and the dendrogram itself. Let's go through each method with examples using a real dataset

- 1. Elbow Method
The elbow method looks for the "elbow point" in a plot of the within-cluster sum of squares (WCSS) or other clustering criteria against the number of clusters. The elbow point is where the rate of decrease in the criterion sharply changes, indicating an optimal number of clusters.
